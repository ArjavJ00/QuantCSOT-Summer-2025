```python
tickers = ['aapl', 'msft', 'goog', 'amzn', 'tsla']
folder_path = "C:\\Users\\arjav\\anaconda_projects\\1d88d8c3-b636-45d9-869b-e903020e95a4\\Stocks"  # Change this if your files are in a subfolder
```

```python
dfs = []

for ticker in tickers:
    file_path = os.path.join(folder_path, f"{ticker}.us.txt")
    
    df = pd.read_csv(file_path)
    
    df['Date'] = pd.to_datetime(df['Date'])  # Convert date to datetime
    df['Ticker'] = ticker.upper()  # Add ticker column
    
    df.set_index(['Ticker', 'Date'], inplace=True)  # Set multi-index
    
    dfs.append(df)

# Combine all into one DataFrame
multi_index_df = pd.concat(dfs)
```

```python
multi_index_df.head()
```

```python
# Option A: Use `get_level_values`
tickers_in_df = multi_index_df.index.get_level_values('Ticker').unique()
print("Tickers found in the DataFrame:", tickers_in_df.tolist())
```

```python
# Grab the dtype of the “Date” level of the MultiIndex
dt_dtype = multi_index_df.index.get_level_values('Date').dtype
print("Date index dtype:", dt_dtype)
```

```python
multi_index_df = multi_index_df.sort_index(level=['Ticker', 'Date'])

# Quick check
multi_index_df.head()
```

```python
for ticker in multi_index_df.index.get_level_values('Ticker').unique():
    # 1) Extract the slice for this ticker (drops the outer index level 'Ticker')
    df_t = multi_index_df.xs(ticker, level='Ticker').copy()
    
    # 2) Ensure it's sorted by date
    df_t = df_t.sort_index()
    
    # 3) Create a "complete" date index from first to last date of this ticker
    start_date = df_t.index.min()
    end_date   = df_t.index.max()
    
    # You can choose 'B' for business days or 'D' for calendar days. 
    # Usually, financial data is measured on trading/business days:
    full_idx = pd.date_range(start=start_date, end=end_date, freq='B')
    
    # 4) Reindex to that full index. This inserts rows of all‐NaN for missing dates.
    df_t_full = df_t.reindex(full_idx)
    
    # 5) How many dates were fully missing? (i.e., all columns NaN)
    fully_missing = df_t_full.isna().all(axis=1).sum()
    print(f"{ticker}: {fully_missing} fully‐missing dates (inserted as NaNs).")
    
    # (Optional) If you want to inspect exactly which dates were missing:
    missing_dates = df_t_full[df_t_full.isna().all(axis=1)].index
    # print(ticker, "missing dates:", missing_dates)  # uncomment to see the list
    
    # 6) Now fill those inserted NaN rows. 
    #    Here we do forward‐fill then backward‐fill in time order:
    df_t_full = df_t_full.ffill().bfill()
    
    # 7) Re‐attach the 'Ticker' level, so it becomes a MultiIndex again
    # First, name the current index level if it's not already named
    df_t_full.index.name = 'Date'  # Add this line to ensure the index has a name
    
    df_t_full['Ticker'] = ticker
    df_t_full = df_t_full.set_index('Ticker', append=True).reorder_levels(['Ticker','Date'])
    
    filled_list.append(df_t_full)

# 8) Concatenate all tickers back into one MultiIndex DataFrame
multi_index_df_fully_cleaned = pd.concat(filled_list).sort_index(level=['Ticker','Date'])

# 9) Final sanity‐check: there should be no rows that are fully NaN now
total_fully_nan_after = (multi_index_df_fully_cleaned.isna().all(axis=1)).sum()
print("\nAfter filling, number of fully‐missing rows across all tickers:", total_fully_nan_after)
```

```python
# ───────────────────────────────────────────────────────────────────────────────
# Cell: Check if any rows still have at least one NaN (after the full‐date fill)
# ───────────────────────────────────────────────────────────────────────────────

# 1) Count total NaNs across all columns
total_nans = multi_index_df_fully_cleaned.isna().sum().sum()
print("Total NaN values across entire DataFrame:", total_nans)

# 2) Count how many rows have at least one NaN
rows_with_any_nan = multi_index_df_fully_cleaned.isna().any(axis=1).sum()
print("Number of rows with any NaN remaining:", rows_with_any_nan)

# 3) (Optional) See which tickers still have any NaNs column‐wise
nan_by_ticker = multi_index_df_fully_cleaned.isna().groupby(level='Ticker').any()
print("\nTicker‐wise “any NaN?”")
display(nan_by_ticker)
```

```python
# ───────────────────────────────────────────────────────────────────────────────
# Cell: Filter to the last 10 years relative to the dataset’s maximum date
# ───────────────────────────────────────────────────────────────────────────────

# 1) Find the maximum date in the fully‐cleaned DataFrame
max_date = multi_index_df_fully_cleaned.index.get_level_values('Date').max()
print("Maximum date in the dataset:", max_date.date())

# 2) Compute the cutoff of “ten years before” that date
ten_years_before = max_date - pd.DateOffset(years=10)
print("Filtering to dates from:", ten_years_before.date(), "→", max_date.date())

# 3) Build a boolean mask and apply it
mask = multi_index_df_fully_cleaned.index.get_level_values('Date') >= ten_years_before
multi_index_last10 = multi_index_df_fully_cleaned[mask]

# 4) Sanity‐check row counts per ticker in this window
counts_last10 = multi_index_last10.groupby(level='Ticker').size()
print("\nNumber of rows per ticker in the last 10 years:")
display(counts_last10)

# 5) Preview the first few rows to confirm
multi_index_last10.head(8)
```

```python
# ───────────────────────────────────────────────────────────────────────────────
# Cell: Add Daily Return, 7‐day & 30‐day MAs, and 30‐day Rolling Volatility
# ───────────────────────────────────────────────────────────────────────────────

# 1) Make a copy so you don’t overwrite the original
df = multi_index_last10.copy()

# 2) DAILY RETURN: percentage change in closing price, by ticker
df['Daily Return'] = (
    df.groupby(level='Ticker')['Close']
      .pct_change() 
)

# 3) 7‐DAY MOVING AVERAGE (MA7) of closing price
df['MA7'] = (
    df.groupby(level='Ticker')['Close']
      .transform(lambda x: x.rolling(window=7, min_periods=1).mean())
)

# 4) 30‐DAY MOVING AVERAGE (MA30) of closing price
df['MA30'] = (
    df.groupby(level='Ticker')['Close']
      .transform(lambda x: x.rolling(window=30, min_periods=1).mean())
)

# 5) ROLLING VOLATILITY (30d): std dev of Daily Return over the last 30 trading days
df['Volatility30'] = (
    df.groupby(level='Ticker')['Daily Return']
      .transform(lambda x: x.rolling(window=30, min_periods=1).std())
)

# 6) Quick sanity‐check: show the first few rows for each ticker
for t in df.index.get_level_values('Ticker').unique():
    print(f"\n--- {t} ---")
    display(df.xs(t, level='Ticker').head(5))

# 7) (Optional) Show a combined head to see all columns together
print("\nCombined DataFrame with new columns:")
df.head(8)
df.columns.tolist()
```

```python
# (It must have a MultiIndex ['Ticker','Date'], plus the column 'Daily Return'.)

# ───────────────────────────────────────────────────────────────────────────────
# Part A: Which stock had the highest average daily return over the period?
# ───────────────────────────────────────────────────────────────────────────────

#  A.1) Compute average daily return per ticker:
avg_returns = df.groupby(level='Ticker')['Daily Return'].mean()

#  A.2) Sort and display:
avg_returns_sorted = avg_returns.sort_values(ascending=False)
print("Average Daily Return per Ticker (10-year window):\n")
display(avg_returns_sorted)

#  A.3) Identify the top‐performer:
top_ticker = avg_returns_sorted.index[0]
top_return = avg_returns_sorted.iloc[0]
print(f"\n→ Highest average daily return: {top_ticker}  ({top_return:.6f} ≈ {top_return*100:.4f}% per day)\n")

# ───────────────────────────────────────────────────────────────────────────────
# Part B: Which stock had the single most volatile calendar month (highest std of daily returns)?
# ───────────────────────────────────────────────────────────────────────────────

#  B.1) Flatten the MultiIndex so that we have Ticker and Date as columns
df_reset = df.reset_index()  # columns now: ['Ticker', 'Date', 'Open', 'High', ..., 'Daily Return', ...]

#  B.2) Create a “YearMonth” column (e.g., 2017-03, 2017-04, etc.)
df_reset['YearMonth'] = df_reset['Date'].dt.to_period('M')

#  B.3) Group by Ticker & YearMonth, compute std‐dev of DAILY RETURN
monthly_vol = (
    df_reset
    .groupby(['Ticker', 'YearMonth'])['Daily Return']
    .std()
    .dropna()
)

#  B.4) Sort and display the top few
monthly_vol_sorted = monthly_vol.sort_values(ascending=False)
print("Top 5 Month‐Ticker Volatilities (std of daily returns):\n")
display(monthly_vol_sorted.head(5))

#  B.5) Pick the single most volatile (ticker, YearMonth) pair
most_volatile_ticker, most_volatile_period = monthly_vol_sorted.index[0]
most_volatile_value = monthly_vol_sorted.iloc[0]
print(f"\n→ Most volatile month: {most_volatile_ticker} in {most_volatile_period}  "
      f"(std dev = {most_volatile_value:.6f} ≈ {(most_volatile_value*100):.4f}%)\n")
```
